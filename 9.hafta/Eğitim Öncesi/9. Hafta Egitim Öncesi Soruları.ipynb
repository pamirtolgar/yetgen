# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
# 1.Problemin Çözümünü Buraya Yazınız

class Vehicle:
    def __init__(self, name, speed, mileage):
        self.name = name
        self.speed = speed
        self.mileage = mileage
        
class Bus(Vehicle):
    def seating_capacity(self, capacity=50):
        print("The seating capacity of a bus is {} passengers".format(capacity))
        
bus = Bus("School Volvo", 180, 12)
bus.seating_capacity()
````

School_bus'ın aynı zamanda Vehicle sınıfının bir örneği olup olmadığını belirleyiniz.

````python
# 2.Problemin Çözümünü Buraya Yazınız

# Vehicle sınıfı tanımı
class Vehicle:
    def __init__(self, name, max_speed, mileage):
        self.name = name
        self.max_speed = max_speed
        self.mileage = mileage

# Bus sınıfı tanımı
class Bus(Vehicle):
    def __init__(self, name, max_speed, mileage):
        super().__init__(name, max_speed, mileage)
        
        self.capacity = 50  # varsayılan kapasite

# School_bus örneğini oluşturma
School_bus = Bus("School Volvo", 180, 12)

# School_bus'ın Vehicle sınıfının bir örneği olup olmadığını kontrol etme
print(isinstance(School_bus, Vehicle))  # True
````
