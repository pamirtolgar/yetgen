```python
# Pizza Sipariş Uygulaması

# Kenarlar
kenarlar = ["ince", "kalın", "kenar yok"]

# Peynir çeşitleri
peynirler = ["mozarella", "cheddar", "parmesan"]

# Fiyatlar
fiyatlar = {"ince": {"mozarella": 20, "cheddar": 22, "parmesan": 24},
            "kalın": {"mozarella": 22, "cheddar": 24, "parmesan": 26},
            "kenar yok": {"mozarella": 18, "cheddar": 20, "parmesan": 22}}

# Kullanıcı Bilgileri
ad = input("Adınız: ")
soyad = input("Soyadınız: ")
adres = input("Adresiniz: ")

# Sipariş Bilgileri
kenar = input("Kenar seçiniz: ")
peynir = input("Peynir seçiniz: ")
adet = int(input("Kaç adet: "))

# Hesaplama
tutar = fiyatlar[kenar][peynir] * adet

# Çıktı
print(f"Sipariş Bilgileri: \n{ad} {soyad} \n{adres} \n{kenar} kenar, {peynir} peynir, {adet} adet. Toplam Tutar: {tutar} TL")
```

```python
# Kullanıcı bilgilerini al
ad = input("İsminiz nedir? ")
soyad = input("Soyisminiz nedir? ")
telefon = input("Telefon numaranız nedir? ")
adres = input("Siparişinizi nereye teslim etmemiz gerekiyor? ")

# Menüyü tanımla
menu = {
  "Margarita": {
    "Malzemeler": ["Domates sosu", "Mozzarella peyniri"],
    "Fiyat": 25
  },
  "Pepperoni": {
    "Malzemeler": ["Domates sosu", "Mozzarella peyniri", "Jambonlu sucuk"],
    "Fiyat": 30
  },
  "Meksika": {
    "Malzemeler": ["Domates sosu", "Mozzarella peyniri", "Jalapeno biberi", "Kırmızı soğan", "Mısır", "Fasulye"],
    "Fiyat": 35
  }
}

# Menüyü ekrana yazdır
print("Menü:\n")
for pizza, detaylar in menu.items():
  print(pizza)
  for malzeme in detaylar["Malzemeler"]:
    print(f"- {malzeme}")
  print(f"Fiyat: {detaylar['Fiyat']} TL\n")

# Sipariş bilgilerini topla
secim = input("Hangi pizzayı seçmek istersiniz? ")
adet = int(input("Kaç adet istersiniz? "))

# Sipariş fiyatını hesapla
toplam_fiyat = menu[secim]["Fiyat"] * adet

# Sipariş bilgilerini ekrana yazdır
print("\nSipariş bilgileri:\n")
print(f"Sipariş veren: {ad} {soyad}")
print(f"Telefon: {telefon}")
print(f"Adres: {adres}\n")
print(f"Sipariş: {secim}")
print(f"Adet: {adet}")
print(f"Toplam fiyat: {toplam_fiyat} TL")

```
